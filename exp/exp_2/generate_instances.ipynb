# create delta

```python
N_p = 20
N_a = 20
Delta_min = 0.2
Delta_max = 1
```

```python
def generate_random_means(N, Delta_min, Delta_max):
    deltas = np.random.uniform(low=Delta_min, high=Delta_max, size=(N-1))
    actual_min_delta = np.min(deltas)
    deltas = np.concatenate([[0], deltas])

    mean_rewards = np.cumsum(deltas)
    np.random.shuffle(mean_rewards)
    return mean_rewards, actual_min_delta
```

```python
def generate_instance(N_p, N_a, Delta_min, Delta_max):
    min_deltas = []
    mean_rewards_p = []
    for p in range(N_p):
        mean_rewards, actual_min_delta = generate_random_means(N_a, Delta_min, Delta_max)
        mean_rewards_p += [mean_rewards]
        min_deltas+= [actual_min_delta]
    mean_rewards_p = np.array(mean_rewards_p)
    
    mean_rewards_a = []
    for a in range(N_a):
        mean_rewards, actual_min_delta = generate_random_means(N_p, Delta_min, Delta_max)
        mean_rewards_a += [mean_rewards]
        min_deltas+= [actual_min_delta]
        
    mean_rewards_a = np.array(mean_rewards_a)
    min_delta = np.min(min_deltas)
    instance = (mean_rewards_p, mean_rewards_a,  N_p, N_a, min_delta)
    return instance
```

```python
instances  = []
for _ in range(100):
    instance = generate_instance(N_p, N_a, Delta_min, Delta_max)
    instances += [instance]
```
